```pyspark
# =============================================================================
# VANILLA LA BATCH PROFILE DETECTION NOTEBOOK
# =============================================================================
# This notebook performs anomaly detection using BATCH PROFILES (organizationId+userId):
# - Loads batch window profiles (14-day merged profiles with organizationId and userId)
# - Joins test data on BOTH organizationId AND userId (org-specific detection)
# - Computes novelty scores using native Spark (no UDFs)
# - Filters for anomalous behavior
# =============================================================================

# Standard library imports
import sys
import os
import logging
import json
import functools
from datetime import datetime, timedelta, date

# PySpark imports
from pyspark.sql import SparkSession, DataFrame
import pyspark.sql.functions as F
from pyspark.sql.types import (
    StructType, StructField, StringType, IntegerType, FloatType, 
    ArrayType, DoubleType, TimestampType, MapType
)


ALLOWLISTED_OPERATORS = [
    "CATO_NETWORK_VPN",
    "CDN_PROXY",
    "CHECKPOINT_VPN",
    "CISCO_UMBRELLA_PROXY",
    "CLOUDCONNEXA_VPN",
    "CYBEROAM_VPN",
    "FORCEPOINT_GATEWAY_PROXY",
    "FORTICLIENT_VPN",
    "GENYMOTION_VDI",
    "GLOBAL_PROTECT_CLOUD_VPN",
    "GOODACCESS_VPN",
    "GOOGLE_ONE_VPN",
    "GOOGLE_RENDER_PROXY",
    "GOOGLE_VPN",
    "HEXTECH_PROXY",
    "IBOSS_VPN",
    "ICLOUD_RELAY_PROXY",
    "JAMF_CONNECT_PROXY",
    "KNOWN_CORPORATE_VPN",
    "MCAFEE_WGCS_VPN",
    "MENLO_SECURITY_VPN",
    "NETSKOPE_PROXY",
    "NORDLAYER_VPN",
    "NSOF_NETWORK_PROXY",
    "PAPERSPACE_VDI",
    "PEPLINK_CLOUD_VPN",
    "PERIMETER81_VPN",
    "QIANXIN_VPN",
    "SONICWALL_VPN",
    "SYMANTEC_WSS_VPN",
    "VIGOR_SSL_VPN",
    "VIPRE_ENCRYPT_VPN",
    "WARP_VPN",
    "ZSCALER_PROXY",
]

ALLOWED_USER_TYPES = [
    "standard",
    "partner",
    "powerpartner",
    "customersuccess",
    "licensemanager",
    "sfdcadmin",
    "externalwho",
    "powercustomersuccess",
]

# VPN operators not marked anonymous in spur but still suspicious
BLOCKLISTED_OPERATORS = [
    "NSOCKS",
    "TOR_PROXY",
    "MULLVAD_VPN",
    "IPROYAL_PROXY",
    "PIA_VPN",
    "CYBER_GHOST_VPN",
    "ASTRILL_VPN",
]

# Configure logging
logger = logging.getLogger("vanilla_LA_batch_detection")
logging.basicConfig(
    stream=sys.stdout,
    format="%(asctime)s.%(msecs)d %(name)s %(levelname)s %(message)s",
    datefmt="%H:%M:%S",
    level=logging.INFO,
)
```

```pyspark
# =============================================================================
# DATA PREPROCESSING FUNCTIONS
# =============================================================================

def la_preprocessing_OPT(df):
    """
    Preprocess the input DataFrame by extracting features from user agent and timestamp.
    Uses native Spark functions for maximum performance.

    Args:
        df (DataFrame): Input DataFrame with browserType, sourceIp, timestamp, loginSubType

    Returns:
        DataFrame with additional columns: browser, os, device, dayofweek, hourofday, ip_octet, loginSubType_prefix
    """
    # Extract primary token (first word after removing Mozilla prefix)
    primary_token = F.regexp_extract(
        F.regexp_replace(F.col("browserType"), r"^Mozilla/\d+\.\d+\s+", ""),
        r"([A-Za-z][A-Za-z0-9._-]*)", 1
    )
    df = df.withColumn("primary_token_norm", F.initcap(primary_token))

    # Browser detection with enhanced logic
    # Order matters: check specific browsers before generic "Chrome" since
    # Edge, SamsungBrowser UAs contain "Chrome"
    browser_col = (
        F.when(F.col("browserType").contains("Edg"), "Edge")
        .when(F.col("browserType").contains("SamsungBrowser"), "SamsungBrowser")
        .when(F.col("browserType").contains("Chrome"), "Chrome")
        .when(F.col("browserType").contains("Firefox"), "Firefox")
        .when(
            F.col("browserType").contains("Safari")
            & ~F.col("browserType").contains("Chrome"),
            "Safari",
        )
        .when(
            F.col("browserType").rlike(r"^[a-zA-Z][a-zA-Z0-9._-]*/[0-9]")
            & ~F.col("browserType").startswith("Mozilla"),
            F.regexp_extract(F.col("browserType"), r"^([a-zA-Z][a-zA-Z0-9._-]*)/", 1),
        )
        .when(
            F.col("primary_token_norm").isNotNull()
            & (F.length(F.col("primary_token_norm")) > 0),
            F.col("primary_token_norm"),
        )
        .otherwise("Unknown")
    )

    # OS parsing
    os_col = (
        F.when(F.col("browserType").contains("Windows"), "Windows")
        .when(
            F.col("browserType").contains("iPhone")
            | F.col("browserType").contains("iPad")
            | F.col("browserType").contains("iOS"),
            "iOS",
        )
        .when(
            F.col("browserType").contains("Macintosh")
            | F.col("browserType").contains("Mac OS"),
            "macOS",
        )
        .when(F.col("browserType").contains("Android"), "Android")
        .when(F.col("browserType").contains("Linux"), "Linux")
        .otherwise("Unknown")
    )

    # Device detection - check Tablet BEFORE Mobile since Android tablets contain both
    device_col = (
        F.when(
            F.col("browserType").contains("Tablet") | F.col("browserType").contains("iPad"),
            "Tablet",
        )
        .when(
            F.col("browserType").contains("Mobile")
            | F.col("browserType").contains("iPhone")
            | F.col("browserType").contains("Android"),
            "Mobile",
        )
        .otherwise("Desktop")
    )

    # Timestamp parsing
    ts_col = F.to_timestamp(F.col("timestamp"))
    
    # Extract day of week and hour of day
    spark_dayofweek = F.dayofweek(ts_col)
    dayofweek_col = F.when(spark_dayofweek == 1, 7).otherwise(spark_dayofweek - 1)
    hourofday_col = F.hour(ts_col)
    
    # Apply all transformations
    processed_df = (
        df.withColumn("browser", browser_col)
        .withColumn("os", os_col)
        .withColumn("device", device_col)
        .withColumn("dayofweek", dayofweek_col)
        .withColumn("hourofday", hourofday_col)
        .withColumn("ip_octet", F.regexp_extract(F.col("sourceIp"), r"^(\d+\.\d+)", 1))
        .withColumn("loginSubType_prefix", F.when(F.col("loginSubType").contains("oauth"), "oauth").otherwise("non_oauth"))
        .drop("primary_token_norm")
    )

    print("la_preprocessing: Native Spark SQL optimization complete.")
    return processed_df
```

```pyspark
# =============================================================================
# ADDITIONAL FILTERING FUNCTIONS
# =============================================================================

def spur_integration(scored_df, spur_path, spark, run_date):
    """
    Enrich detections with SPUR data and compute confidence levels.

    Confidence level logic:
      - If operator IN BLOCKLISTED_OPERATORS -> "blocklisted" (always suspicious)
      - If anonymous=True AND operator NOT in allowlist -> "critical"
      - If tunnels exist AND ALL operators are allowlisted -> "drop"
      - Otherwise (no spur match or no tunnels) -> "high-tier"

    Returns events with confidence_level in ("critical", "blocklisted").

    Args:
        scored_df: DataFrame with anomaly scores
        spur_path: Path to SPUR data
        spark: SparkSession instance
        run_date: Current run date

    Returns:
        DataFrame: Critical + blocklisted detections enriched with SPUR data
    """
    try:
        df_spur = spark.read.parquet(spur_path)
    except Exception as e:
        print(f"Warning: Could not load SPUR data for {run_date}: {str(e)}")
        print("Returning empty DataFrame (no spur data means no critical events)")
        return scored_df.limit(0)

    rawmsg_schema = StructType([
        StructField("ip", StringType()),
        StructField("organization", StringType()),
        StructField("as", MapType(StringType(), StringType())),
        StructField("client", StructType([
            StructField("proxies", ArrayType(StringType())),
            StructField("behaviors", ArrayType(StringType())),
        ])),
        StructField("location", StructType([
            StructField("city", StringType()),
            StructField("state", StringType()),
            StructField("country", StringType())
        ])),
        StructField("risks", ArrayType(StringType()))
    ])

    df_parsed = df_spur.withColumn(
        "RawMsg_json",
        F.from_json(F.col("RawMsg"), rawmsg_schema)
    )

    # Build compact per-IP spur classification once, then broadcast-join (faster than joining all spur rows)
    df_selected = df_parsed.select(
        F.col("RawMsg_json.ip").alias("ip"),
        F.col("tunnels").alias("tunnels"),
        F.coalesce(F.col("RawMsg_json.client.proxies"), F.array()).alias("client_proxies"),
        F.coalesce(F.col("RawMsg_json.client.behaviors"), F.array()).alias("client_behaviors"),
        F.coalesce(F.col("RawMsg_json.risks"), F.array()).alias("risks"),
        F.coalesce(F.transform(F.col("tunnels"), lambda x: x["operator"]), F.array()).alias("tunnel_operators"),
    ).filter(F.col("ip").isNotNull())

    allowlist_str = ",".join([f"'{op}'" for op in ALLOWLISTED_OPERATORS])
    blocklist_tokens_array = "array(" + ",".join([f"'{op.lower()}'" for op in BLOCKLISTED_OPERATORS]) + ")"

    has_tunnels_expr = F.col("tunnels").isNotNull() & (F.size(F.col("tunnels")) > 0)
    has_non_allowlisted_expr = F.expr(
        f"exists(tunnels, x -> x.operator IS NULL OR x.operator NOT IN ({allowlist_str}))"
    )
    has_anon_non_allowlisted_expr = F.expr(
        f"exists(tunnels, x -> x.anonymous = true AND (x.operator IS NULL OR x.operator NOT IN ({allowlist_str})))"
    )

    # Build a unified set of SPUR signals and match blocklist tokens case-insensitively.
    all_spur_signals_expr = F.expr(
        "array_distinct(flatten(array(client_proxies, client_behaviors, risks, tunnel_operators)))"
    )
    matched_blocklist_proxies_expr = F.expr(
        f"filter({blocklist_tokens_array}, tok -> exists(array_distinct(flatten(array(client_proxies, client_behaviors, risks, tunnel_operators))), s -> lower(s) LIKE concat('%', tok, '%')))"
    )
    has_blocklisted_expr = F.size(matched_blocklist_proxies_expr) > 0

    # Compute numeric priority first, then aggregate by IP (avoids lexical max on strings)
    # Keep matched_proxies so downstream detections can be filtered by proxy type.
    per_ip = (
        df_selected
        .withColumn("matched_blocklist_proxies", matched_blocklist_proxies_expr)
        .withColumn(
            "priority",
            F.when(has_blocklisted_expr, 4)
            .when(has_anon_non_allowlisted_expr, 3)
            .when(has_tunnels_expr & ~has_non_allowlisted_expr, 2)
            .otherwise(1)
        )
        .groupBy("ip")
        .agg(
            F.max("priority").alias("priority"),
            F.array_distinct(F.flatten(F.collect_list("matched_blocklist_proxies"))).alias("matched_proxies"),
        )
        .withColumn(
            "confidence_level",
            F.when(F.col("priority") == 4, "blocklisted")
            .when(F.col("priority") == 3, "critical")
            .when(F.col("priority") == 2, "drop")
            .otherwise("high-tier")
        )
        .drop("priority")
    )

    joined_df = scored_df.join(F.broadcast(per_ip), scored_df.sourceIp == per_ip.ip, how="left").drop("ip")
    joined_df = joined_df.localCheckpoint()
    joined_df = joined_df.withColumn("confidence_level", F.coalesce(F.col("confidence_level"), F.lit("high-tier")))

    summary = joined_df.groupBy("confidence_level").count()
    total = joined_df.count()
    b = summary.filter(F.col("confidence_level") == "blocklisted").select(F.coalesce(F.first("count"), F.lit(0))).collect()[0][0]
    c = summary.filter(F.col("confidence_level") == "critical").select(F.coalesce(F.first("count"), F.lit(0))).collect()[0][0]
    h = summary.filter(F.col("confidence_level") == "high-tier").select(F.coalesce(F.first("count"), F.lit(0))).collect()[0][0]
    d = summary.filter(F.col("confidence_level") == "drop").select(F.coalesce(F.first("count"), F.lit(0))).collect()[0][0]

    print("Confidence level breakdown:")
    print(f"   Total anomalies: {total:,}")
    print(f"   Blocklisted (emitted): {b:,}")
    print(f"   Critical (emitted): {c:,}")
    print(f"   High-tier (suppressed): {h:,}")
    print(f"   Drop (suppressed): {d:,}")

    return joined_df.filter(F.col("confidence_level").isin("critical", "blocklisted"))
```

```pyspark
# =============================================================================
# NOVELTY DETECTION FUNCTIONS - batch PROFILES
# =============================================================================

def get_novelty_scores_batch_profile(test_df, profile_df):
    """
    Compute novelty scores using batch PROFILES (organizationId + userId).
    
    Key differences from single userId approach:
    - Joins on BOTH organizationId AND userId (org-specific detection)
    - Uses batch window profiles with separate organizationId and userId columns
    - Otherwise, same native Spark novelty computation logic
    
    Args:
        test_df: Test DataFrame with organizationId, userId, and feature columns
        profile_df: batch window profile DataFrame with organizationId, userId, and profile columns
        
    Returns:
        DataFrame with novelty scores for anomalous rows only
    """
    logger.info("Using batch profile (organizationId+userId) novelty detection with inline filtering!")
    
    # No hard filtering — compute novelty scores for ALL features, filter downstream per path
    filtering_features = []
    non_filtering_features = [
        "ip_octet", "sourceIp", "browser", "os",
        "countryCode", "loginType",
        "loginSubType_prefix", "device", "dayofweek", "hourofday"
    ]
    
    all_features = filtering_features + non_filtering_features
    print(f"Filtering features (will inline filter): {filtering_features}")
    print(f"Non-filtering features (score only): {non_filtering_features}")
    
    # batch PROFILE JOIN: Join on BOTH organizationId AND userId
    print("\n--- Performing BATCH PROFILE JOIN (organizationId + userId) ---")
    profile_df_cached = profile_df.repartition(200, F.col("organizationId"), F.col("userId")).cache()
    test_df_partitioned = test_df.repartition(200, F.col("organizationId"), F.col("userId"))
    
    # FIX: Handle duplicate column names properly during JOIN
    # Get columns from each DataFrame, excluding join keys and common columns
    test_df_cols = [c for c in test_df_partitioned.columns if c not in ["organizationId", "userId"]]
    profile_df_cols = [c for c in profile_df_cached.columns if c not in ["organizationId", "userId"]]
    
    # Find common columns (excluding join keys) that would cause ambiguity
    common_cols = set(test_df_cols) & set(profile_df_cols)
    if common_cols:
        print(f"FIX: Found common columns that could cause ambiguity: {common_cols}")
        # For common columns, take from test_df and exclude from profile_df
        profile_df_cols = [c for c in profile_df_cols if c not in common_cols]
    
    joined_df = test_df_partitioned.join(
        profile_df_cached, 
        (test_df_partitioned.organizationId == profile_df_cached.organizationId) & 
        (test_df_partitioned.userId == profile_df_cached.userId),
        how="inner"
    ).select(
        test_df_partitioned.organizationId.alias("organizationId"),
        test_df_partitioned.userId.alias("userId"),
        *[test_df_partitioned[c] for c in test_df_cols],
        *[profile_df_cached[c] for c in profile_df_cols]
    )
    
    # FILTER OUT ZERO-COUNT PROFILES: Exclude events with zero counts
    print(f"\nFiltering out zero-count profiles...")
    initial_count = joined_df.count()
    print(f"Starting with {initial_count:,} rows after INNER JOIN")
    
    # Filter out zero-count profiles for all features
    filtered_df = joined_df
    for feature in filtering_features:
        total_col = f"{feature}_total_count"
        filtered_df = filtered_df.filter(F.col(total_col) > 0)
    
    filtered_count = filtered_df.count()
    print(f"After filtering zero-count profiles: {filtered_count:,} rows")
    print(f"Excluded {initial_count - filtered_count:,} events with zero counts")
    
    # INLINE FILTERING: Only process filtering features, filter as we go
    result_df = filtered_df
    
    # Process filtering features with inline filtering
    for i, feature in enumerate(filtering_features):
        profile_col = f"{feature}_profile"
        total_col = f"{feature}_total_count"
        
        print(f"\nProcessing filtering feature {i+1}/{len(filtering_features)}: {feature}")

        
        # Check if profile maps are empty
        empty_profiles = result_df.filter(
            F.col(profile_col).isNull() | 
            (F.size(F.col(profile_col)) == 0)
        ).count()
        print(f"   DEBUG: {empty_profiles:,} rows have empty {profile_col} maps")
        
        # INLINE FILTERING: Keep only rows where novelty == 1.0
        result_df = result_df.filter(
            # Compute novelty and check if == 1.0
            (1.0 - (
                F.coalesce(
                    F.col(profile_col).getItem(F.col(feature).cast("string")), 
                    F.lit(0)
                ) / F.col(total_col).cast("double")
            )) == 1.0
        )
        
        # Track progress
        remaining_count = result_df.count()
        reduction_pct = (1 - remaining_count/filtered_count) * 100 if filtered_count > 0 else 0
        print(f"   After {feature} inline filter: {remaining_count:,} rows ({reduction_pct:.1f}% total reduction)")
        
        # Early termination if no rows left
        if remaining_count == 0:
            print("Early termination: No anomalous rows found")
            break
            
        # Smart caching after major reductions
        if remaining_count < filtered_count * 0.1 and i == 1:
            print(f"Caching after major reduction to {remaining_count:,} rows")
            result_df = result_df.cache()
    
    # Add novelty=1.0 columns for filtering features (all remaining rows are anomalous)
    print(f"\nAdding novelty=1.0 columns for {len(filtering_features)} filtering features")
    for feature in filtering_features:
        result_df = result_df.withColumn(f"{feature}_novelty", F.lit(1.0))
    
    # Compute scores for non-filtering features (but don't filter)
    print(f"Computing scores for {len(non_filtering_features)} non-filtering features")
    for feature in non_filtering_features:
        profile_col = f"{feature}_profile"
        total_col = f"{feature}_total_count"
        print(f"   Computing score for {feature}")
        result_df = result_df.withColumn(
            f"{feature}_novelty",
            F.when(
                F.col(profile_col).isNull() |
                (F.size(F.col(profile_col)) == 0) |
                F.col(total_col).isNull() |
                (F.col(total_col) <= 0),
                F.lit(0.0)
            ).otherwise(
                1.0 - (
                    F.coalesce(
                        F.col(profile_col).getItem(F.col(feature).cast("string")),
                        F.lit(0)
                    ) / F.col(total_col).cast("double")
                )
            )
        )
    
    # Select final columns (including datacenter and timestamp)
    feature_columns = ["userId", "organizationId"] + all_features
    novelty_columns = [f"{feature}_novelty" for feature in all_features]
    total_count_columns = [f"{feature}_total_count" for feature in all_features]

    # Add metadata columns if they exist in result_df
    metadata_columns = []
    if "datacenter" in result_df.columns:
        metadata_columns.append("datacenter")
    if "timestamp" in result_df.columns:
        metadata_columns.append("timestamp")
    if "rootRequestId" in result_df.columns:
        metadata_columns.append("rootRequestId")

    # Select all columns: features, novelty scores, total counts, and metadata
    final_result = result_df.select(*(feature_columns + novelty_columns + total_count_columns + metadata_columns))
    
    final_count = final_result.count()
    print(f"\n BATCH profile detection complete!")
    print(f"Final result: {final_count:,} anomalous rows ({(final_count/filtered_count)*100:.2f}% of filtered data)")
    if metadata_columns:
        print(f"Metadata columns included: {metadata_columns}")
    
    profile_df_cached.unpersist()
    return final_result
```

```pyspark
# =============================================================================
# PARAMETERIZED DETECTION PIPELINE
# =============================================================================
#
# Two detection paths, each with its own novelty gate:
#   Path 1 (blocklisted proxy): variable novelty features (passed as param)
#   Path 2 (anonymous proxy):   fixed 6-feature novelty (ip_octet, sourceIp, browser, os, loginType, countryCode)
#
# The novelty gate runs AFTER spur classification so each path can have different features.

TRUSTED_IP_PREFIXES = ["13.110.24", "13.108"]
ANONYMOUS_NOVELTY_FEATURES = ["ip_octet", "sourceIp", "browser", "os", "loginType", "countryCode"]


def handle_detection_batch(test_df, profile_df, spur_path, spark, run_date,
                           blocklist_novelty_features=None,
                           use_strict_prefilter=False):
    """
    Parameterized detection pipeline.

    Args:
        test_df: Raw login event DataFrame (already column-renamed)
        profile_df: Batch rolling profile DataFrame
        spur_path: Path to SPUR parquet data
        spark: SparkSession
        run_date: Date being processed
        blocklist_novelty_features: List of features that must be novel for blocklisted path
                                    (e.g. ["ip_octet", "sourceIp", "os"])
        use_strict_prefilter: If True, common pre-filter requires
                             sourceIp/ip_octet/os/browser novelty == 1.0

    Returns:
        DataFrame of detections
    """
    if blocklist_novelty_features is None:
        blocklist_novelty_features = ["ip_octet", "sourceIp", "browser", "os"]

    print(f"Config:")
    print(f"  Blocklist novelty: {blocklist_novelty_features}")
    print(f"  Anonymous novelty: {ANONYMOUS_NOVELTY_FEATURES}")
    print(f"  Strict common pre-filter (ip_octet, sourceIp, os, browser): {use_strict_prefilter}")

    # --- Step 1: Filter trusted IPs ---
    trusted_filter = F.lit(False)
    for prefix in TRUSTED_IP_PREFIXES:
        trusted_filter = trusted_filter | F.col("sourceIp").startswith(prefix)
    filtered_df = test_df.filter(~trusted_filter)

    pre_count = test_df.count()
    post_trusted = filtered_df.count()
    print(f"Step 1 - Trusted IP filter: {pre_count:,} -> {post_trusted:,} "
          f"(removed {pre_count - post_trusted:,})")

    # --- Step 2: Filter standard users ---
    filtered_df = filtered_df.filter(
        F.lower(F.col("userType")).isin(ALLOWED_USER_TYPES)
    )
    post_user = filtered_df.count()
    print(f"Step 2 - User type filter: {post_trusted:,} -> {post_user:,} "
          f"(removed {post_trusted - post_user:,})")

    # --- Step 3: Preprocessing ---
    processed_df = la_preprocessing_OPT(filtered_df)
    pre_bot = processed_df.count()
    processed_df = processed_df.filter(F.col("browser") != "Slackbot")
    print(f"Step 3 - Preprocessing complete: {pre_bot:,} rows, after Slackbot filter: {processed_df.count():,}")

    # --- Step 4: Compute all novelty scores (no hard filter) ---
    print(f"Step 4 - Computing novelty scores (no hard filter)...")
    scored_df = get_novelty_scores_batch_profile(processed_df, profile_df)

    if scored_df is None:
        print("No events after profile join.")
        return scored_df

    prefilter_expr = (
        (F.col("sourceIp_novelty") == 1.0)
        & (F.col("ip_octet_novelty") == 1.0)
        & (F.col("os_novelty") == 1.0)
    )
    
   

    # Cache once: reused by count + spur join
    scored_df = scored_df.filter(prefilter_expr).localCheckpoint()
    scored_count = scored_df.count()
    if scored_count == 0:
        print("No events after common novelty pre-filter.")
        scored_df.unpersist()
        return scored_df.limit(0)

    print(f"Step 4 - After common novelty pre-filter: {scored_count:,}")

    # --- Step 5: Spur integration ---
    print(f"Step 5 - Spur integration...")
    spur_df = spur_integration(scored_df, spur_path, spark, run_date="2026-03-17")

    if spur_df is None:
        scored_df.unpersist()
        print("No spur-matched events found. No detections.")
        return spur_df

    spur_count = spur_df.count()
    if spur_count == 0:
        scored_df.unpersist()
        print("No spur-matched events found. No detections.")
        return spur_df

    # --- Step 6: Two detection paths with different novelty gates ---
    # Path 1: Blocklisted → apply blocklist_novelty_features
    blocklisted = spur_df.filter(F.col("confidence_level") == "blocklisted")
    for feat in blocklist_novelty_features:
        blocklisted = blocklisted.filter(F.col(f"{feat}_novelty") == 1.0)

    # Path 2: Critical (anonymous proxy) → apply fixed 6-feature novelty
    critical = spur_df.filter(F.col("confidence_level") == "critical")
    for feat in ANONYMOUS_NOVELTY_FEATURES:
        critical = critical.filter(F.col(f"{feat}_novelty") == 1.0)

    detections = blocklisted.unionByName(critical)

    p1, p2 = blocklisted.count(), critical.count()
    print(f"\n   Path 1 (blocklisted + {blocklist_novelty_features}): {p1:,}")
    print(f"   Path 2 (anon proxy + {ANONYMOUS_NOVELTY_FEATURES}):   {p2:,}")
    print(f"   Total detections: {p1 + p2:,}")

    scored_df.unpersist()
    return detections


def run_full_pipeline(
    input_path, window_profile_path, spur_path, spark, run_date,
    output_path=None, blocklist_novelty_features=None, use_strict_prefilter=False
):
    """
    Load data, run full detection pipeline, optionally save results.

    Args:
        input_path: S3 path to applog parquet data
        window_profile_path: S3 path to rolling batch profiles
        spur_path: S3 path to spur parquet data
        spark: SparkSession
        run_date: Date being processed
        output_path: Optional S3 path to write critical detections
        blocklist_novelty_features: List of features for blocklist novelty gate
        use_strict_prefilter: If True, common pre-filter includes os_novelty and browser_novelty

    Returns:
        DataFrame of detections
    """
    spark.conf.set("spark.sql.optimizer.dynamicPartitionPruning.enabled", "false")
    spark.conf.set("spark.sql.adaptive.reuseSubquery.enabled", "false")

    # Load and deduplicate test data
    df_test = spark.read.parquet(input_path).dropDuplicates()
    df_test = df_test.filter(
        ~(
            (F.col("userId_basic") == "000000000000000")
            | (F.col("organizationId_basic") == "000000000000000")
            | (F.col("resultCode_L") != 0)
            | (F.col("userType_common") == "Guest")
            | (F.trim(F.col("loginUrl_L")) == "test.salesforce.com")
            | (F.trim(F.col("loginUrl_L")).contains(".sandbox."))
        )
    )
    df_test = df_test.filter(
        F.lower(F.col("userType_common")).isin(ALLOWED_USER_TYPES)
    )

    # Column renaming
    df_selected = df_test.select(
        F.col("organizationId_basic").alias("organizationId"),
        F.col("userId_basic").alias("userId"),
        F.col("browserType_L").alias("browserType"),
        F.col("timestamp").alias("timestamp"),
        F.col("sourceIp_L").alias("sourceIp"),
        F.col("loginType_L").alias("loginType"),
        F.col("sourcetype").alias("sourcetype"),
        F.col("resultCode_L").alias("resultCode"),
        F.col("environment").alias("environment"),
        F.col("logRecordType").alias("logRecordType"),
        F.col("sidHash").alias("sidHash"),
        F.col("loginSubType_L").alias("loginSubType"),
        F.col("countryCode_L").alias("countryCode"),
        F.col("datacenter").alias("datacenter"),
        F.col("rootRequestId").alias("rootRequestId"),
        F.col("userType_common").alias("userType"),
    )
    df_selected = df_selected.localCheckpoint()


    print(f"Loaded {df_selected.count():,} records from {input_path}")

    # Load profiles
    profile_df = spark.read.parquet(window_profile_path)
    if "orgId" in profile_df.columns:
        profile_df = profile_df.withColumnRenamed("orgId", "organizationId")
    print(f"Loaded {profile_df.count():,} profiles from {window_profile_path}")

    # Run full pipeline
    critical_df = handle_detection_batch(
        df_selected, profile_df, spur_path, spark, run_date,
        blocklist_novelty_features=blocklist_novelty_features,
        use_strict_prefilter=use_strict_prefilter,
    )

    critical_count = 0
    if critical_df is not None:
        critical_count = critical_df.count()

    if output_path and critical_df is not None:
        critical_df.write.mode("overwrite").parquet(output_path)
        print(f"Saved {critical_count:,} critical detections to {output_path}")

    # Datacenter stats
    if critical_df is not None and critical_count > 0:
        print("\n--- Datacenter-wise critical detection stats ---")
        critical_df.groupBy("datacenter").agg(
            F.count("*").alias("critical_count"),
            F.countDistinct("organizationId").alias("unique_orgs"),
            F.countDistinct("userId").alias("unique_users"),
            F.countDistinct("sourceIp").alias("unique_ips"),
        ).orderBy(F.desc("critical_count")).show(50, truncate=False)

    return critical_df
```
